### Step 1: Search Conversations
Search across all the conversations using the following fields and filters:
 - Terms related to the conversation
 - Agent id
 - Customer id
 - Resolved or not
 - Sentiment
 - Rating

The results will bring a summary, grounding and snippets
 - Follow up questions


```python
import p5_contact_center_analyst
```

```python
data = p5_contact_center_analyst.SearchConversationsRequest(
    query="process to return a product",
    user_pseudo_id="renatoleite@google.com",
    status="resolved"
)
responses = p5_contact_center_analyst.search_conversations(data)
```

```python
responses.responses
```

```python
data = p5_contact_center_analyst.SearchConversationsRequest(
    query="for chairs",
    user_pseudo_id="renatoleite@google.com",
    conversation_id=responses.conversation_id)
results = p5_contact_center_analyst.search_conversations(data)
results.responses
```

```python
conversation = results.responses["search_results"][0]["conversation"]
```

```python
data = p5_contact_center_analyst.VectorFindNeighborRequest(
    input_text=conversation,
    user_journey="conversations"
)
results = p5_contact_center_analyst.vector_find_similar(
    data=data
)
```
